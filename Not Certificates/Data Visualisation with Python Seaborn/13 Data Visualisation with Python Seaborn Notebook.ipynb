```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
fires = pd.read_csv('fires.csv')
waste = pd.read_csv('waste.csv')
plants = pd.read_csv('plants.csv')
waste['income'] = waste['income'].astype('category').cat.set_categories(['Low', 'Lower Middle', 'Upper Middle', 'High'], ordered=True)
```

```python
%matplotlib inline
plt.rcParams['figure.dpi']=72
```

## Bar chart

```python
sns.barplot(data=waste, x='gdp', y='income', ci=None)
plt.show()
```

## Scatter plot

```python
sns.scatterplot(data=waste, x='glass', y='plastic')
plt.show()
```

## Line chart

```python
sns.lineplot(data=fires, x='year', y='firespots', ci=None)
plt.show()
```

## Histogram

```python
sns.histplot(data=plants, x='Leaf_length')
plt.show()
```

## KDE plot

```python
sns.kdeplot(data=plants, x='Leaf_width', fill=True)
plt.show()
```

## Box plot

```python
sns.boxplot(data=plants, x='Plant_height', y='PH')
plt.show()
```
