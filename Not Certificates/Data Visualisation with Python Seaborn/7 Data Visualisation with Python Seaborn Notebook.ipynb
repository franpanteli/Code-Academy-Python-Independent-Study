```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
fires = pd.read_csv('fires.csv')
```

```python
%matplotlib inline
plt.rcParams['figure.dpi']=72
```

Examine the first few rows of the dataset.

```python
fires.head()
```

Explore a basic line plot of average firespots by month.

```python
sns.lineplot(data=fires, x='month', y='firespots', ci=None)
plt.show()
```

Subset the data to look at specific states and years for July through December.

```python
fires_sub = fires[fires['state'].isin(['AMAZONAS', 'PARA', 'RONDONIA']) & 
                  fires['year'].isin([1999,2009,2019]) & 
                  fires['month'].isin(list(range(7,13)))]
```

Use seaborn parameters to display three plots in one frame with titles and labels.

```python
sns.set_theme(font_scale=1.3)
fireplot = sns.relplot(kind='line', data=fires_sub, x='month', y='firespots', col='state', hue='year', style='year', palette='colorblind', linewidth=3)
fireplot.set_titles('{col_name}', size=16)
fireplot.fig.suptitle('Regional Fires by Month from 1999 to 2019', size=18, y=1.1)
plt.show()
```
